# 2D-Random Walks
#### our model is simple Random walk implementation in a 2D coordinates , our object have four choices (0,1) , (0,-1) , (1,0) , (-1,0)
#### meaning that our object is moving right or left or up or down in one move.

for example in this situation we imagine that we start from (0,0) and we have four choices to move and suppose we choose (1,0) and after that we have another four choices to make our goal is to understand this randomness in code enviroment

```python
from random import choice
import numpy as np
import matplotlib.pyplot as plt
```

```python
# this code make random choice between what we put inside that
x , y  = 0 , 0
X , Y  = [] , []
for time in range(5900):
    dx , dy  = choice([(1,0) , (-1,0) , (0,1) , (0,-1)])
    x , y   = x+dx , y+dy 
    X.append(x)
    Y.append(y)
```

```python
plt.plot(X, Y)
```
