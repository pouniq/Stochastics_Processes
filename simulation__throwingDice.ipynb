```python
import random
import matplotlib.pyplot as plt

# Settings
rolls_per_trial = 100     # Number of dice rolls per trial
num_trials = 2000        # Number of trials (i.e., repeat the experiment this many times)

# Simulate: Take the average of 'rolls_per_trial' rolls, repeat for 'num_trials'
averages = []
for _ in range(num_trials):
    trial_rolls = [random.randint(1, 6) for _ in range(rolls_per_trial)]
    avg = sum(trial_rolls) / rolls_per_trial
    averages.append(avg)

# Plot the histogram of the averages
plt.figure(figsize=(10, 5))
plt.hist(averages, bins=20, color='skyblue', edgecolor='black', density=True)
plt.title(f'Distribution of Averages from {num_trials} Trials\nEach with {rolls_per_trial} Dice Rolls')
plt.xlabel('Average Dice Roll Value')
plt.ylabel('Probability Density')
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```
